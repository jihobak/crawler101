```python
def get_restaurant_keys(keyword, request_count):
    BASE_URL = "https://stage.mangoplate.com/api/v5/search/by_keyword.json"
    data = {
    "language": "kor",
    "start_index": "0",
    "request_count": request_count,
    "keyword": keyword,
    #"filter": {"subcuisine_codes":[],"metro_codes":[],"price_codes":[],"cuisine_codes":[],"is_parking_available":0},
    "order_by": "2"
    }
    response = requests.post(BASE_URL, data=data)
    response_dict = response.json()
    
    rest_key_store = []
    for res in response_dict['result']:
        rest_key_store.append(res['restaurant']['restaurant_key'])
        
    return rest_key_store
```

```python
def get_comment(rest_key, request_count):
    REVIEW_URL = f"https://stage.mangoplate.com/api/v5/restaurants/{rest_key}/reviews.json"
    params = {
        "language": "kor",
        "start_index": "0",
        "request_count": request_count,
        "sort_by": "2"
    }
    
    response = requests.get(REVIEW_URL, params=params)
    response_dict = response.json()
    
    comment_store = []
    for res in response_dict:
        comment_store.append(res['comment']['comment'])
        
    return comment_store
```

```python
cs = get_comment('9jKcO4Igs-b_', 10)
```

# Flow 

- 레스토랑 키를 가져오는 함수
- 레스토랑 키와 리뷰 수를 주면 리뷰를 가져오는 함수
- 이 두 함수를 활용해서 크롤할 최종'get_reviews' 함수 만들기

```python
def get_reviews(keyword, rest_num):
    rest_key_store = get_restaurant_keys(keyword, rest_num)
    
    review_store = []
    for key in rest_key_store:
        review_store.extend(get_comment(key, 10))
    
    return review_store
```

```python
reviews = get_reviews("대구", 10)
```
